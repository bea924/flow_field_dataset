```python
import matplotlib.pyplot as plt
import pandas as pd
import os

os.makedirs('figures', exist_ok=True)
df= pd.read_csv("figures/losses.csv")
plt.figure(figsize=(6, 5))
plt.plot(df['train_loss'], label='Training Loss', color='blue')
plt.plot(df['validation_loss'], label='Validation Loss', color='orange')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.grid()
plt.savefig('figures/loss_plot.pdf', dpi=300, bbox_inches='tight')
```
